```python
import os

import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from uncertainties import unumpy as unp

from funcs.dir import d_drive
from funcs.post_processing.images.soot_foil.deltas import get_px_deltas_from_lines

sns.set_style("whitegrid")
```

```python
dir_data = os.path.join(
    d_drive,
    "Data",
    "Processed",
    "Soot Foil",
    "foil images",
    "2020-12-09",
    "Shot 00",
)

image = os.path.join(dir_data, "example.png")
mask = os.path.join(dir_data, "example_mask.png")
mm_per_px = 290.0 / 2101.0
```

# No masking

```python
deltas_no_mask = get_px_deltas_from_lines(
    img_path=image,
    mask_path=None,
    use_fast=True,
    apply_uncertainty=False,
) * mm_per_px
```

```python
sns.distplot(deltas_no_mask)
print(np.median(deltas_no_mask))
```

# With masking

```python
deltas_with_mask = get_px_deltas_from_lines(
    img_path=image,
    mask_path=mask,
    use_fast=True,
    apply_uncertainty=False,
) * mm_per_px
```

```python
sns.distplot(deltas_with_mask)
print(np.median(deltas_with_mask))
```
